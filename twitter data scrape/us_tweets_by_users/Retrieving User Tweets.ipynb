```python
import pandas as pd

tweets_df = pd.read_csv("../covid_washington_tweets.csv")
tweets_df.head()
```

```python
df_new = tweets_df[['date', 'renderedContent', 'user']]
df_new.head()
```

```python
# for Quote Network

userlist = ['https://twitter.com/MayorBowser',
            'https://twitter.com/thehill',
            'https://twitter.com/atrupar',
            'https://twitter.com/CNN',
            'https://twitter.com/Heerak4Congress',
            'https://twitter.com/thehill', 
            'https://twitter.com/Dsquared75', 
            'https://twitter.com/washingtonpost',
            'https://twitter.com/realDonaldTrump',
            'https://twitter.com/gtconway3d'
           ]
for user in userlist:
    username = user.split('/')[-1]
    user_tweets = df_new[tweets_df['user'] == user]
    
    user_tweets.to_csv(f'{username}_tweets.csv', index=False)
    user_tweets.to_csv(f'{username}_tweets.txt', index=False, sep='\t')
```
