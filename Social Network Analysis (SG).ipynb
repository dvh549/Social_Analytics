```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
covid_singapore = pd.read_csv("twitter data scrape/covid_singapore_tweets.csv")
covid_singapore.head()
```

```python
coronavirus_singapore = pd.read_csv("twitter data scrape/coronavirus_singapore_tweets.csv")
coronavirus_singapore.head()
```

## Covid

```python
quote_graph = nx.DiGraph()

for i in range(len(covid_singapore)):
    if type(covid_singapore.iloc[i]["quotedTweet"]) == str:
        quote_graph.add_edge(covid_singapore.iloc[i]["user"].split("/")[3], covid_singapore.iloc[i]["quotedTweet"].split("/")[3])
        
# nx.draw(quote_graph)
# nx.write_gexf(quote_graph, "gephi/sg/covid_quote_network.gexf")
```

```python
quoted_betweenness = nx.betweenness_centrality(quote_graph)
quoted_betweenness_desc = dict(sorted(quoted_betweenness.items(), key=lambda item: item[1], reverse=True))
quoted_betweenness_desc
```

```python
quoted_betweenness_cent = pd.DataFrame(list(quoted_betweenness_desc.items())[:10], columns=["User", "Betweenness Centrality"])
quoted_betweenness_cent
```

```python
quoted_closeness = nx.closeness_centrality(quote_graph)
quoted_closeness_desc = dict(sorted(quoted_closeness.items(), key=lambda item: item[1], reverse=True))
quoted_closeness_desc
```

```python
quoted_closeness_cent = pd.DataFrame(list(quoted_closeness_desc.items())[:10], columns=["User", "Closeness Centrality"])
quoted_closeness_cent
```

```python
quoted_in_degree = nx.in_degree_centrality(quote_graph)
quoted_in_degree_desc = dict(sorted(quoted_in_degree.items(), key=lambda item: item[1], reverse=True))
quoted_in_degree_desc
```

```python
quoted_in_degree_cent = pd.DataFrame(list(quoted_in_degree_desc.items())[:10], columns=["User", "In Degree Centrality"])
quoted_in_degree_cent
```

```python
quoted_out_degree = nx.out_degree_centrality(quote_graph)
quoted_out_degree_desc = dict(sorted(quoted_out_degree.items(), key=lambda item: item[1], reverse=True))
quoted_out_degree_desc
```

```python
quoted_out_degree_cent = pd.DataFrame(list(quoted_out_degree_desc.items())[:10], columns=["User", "Out Degree Centrality"])
quoted_out_degree_cent
```
