```python
import pandas as pd

# CSV File can be downloaded using the below link:
# SG - https://smu-my.sharepoint.com/:x:/g/personal/dvhanley_2019_scis_smu_edu_sg/EYUmv94oJD5IkPhkWPU5EKgBpjCUs5DeDCRGqntYUgE6kQ?e=NTWo5a

tweets_df = pd.read_csv("../singapore_tweets.csv")
tweets_df.head()
```

```python
df_new = tweets_df[['date', 'renderedContent', 'user']]
df_new.head()
```

```python
# for Quote Network

userlist = ['https://twitter.com/ChannelNewsAsia', 'https://twitter.com/MothershipSG', 'https://twitter.com/DannyQuah', 'https://twitter.com/BogdziewiczM', 'https://twitter.com/DTrain', 'https://twitter.com/leejanhau', 'https://twitter.com/dwallbank', 'https://twitter.com/temptempholder', 'https://twitter.com/theforceawaken']
for user in userlist:
    username = user.split('/')[-1]
    user_tweets = df_new[tweets_df['user'] == user]
    
    user_tweets.to_csv(f'{username}_tweets.csv', index=False)
    user_tweets.to_csv(f'{username}_tweets.txt', index=False, sep='\t')
```
